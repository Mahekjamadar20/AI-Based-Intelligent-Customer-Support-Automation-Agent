```python
import datetime
import re
import random

class BusinessSupportAgent:

    def __init__(self):
        self.memory = []
        self.ticket_id = 1001


    def greeting_tool(self):
        return random.choice([
            "Hello! How can I assist you today?",
            "Hi there! I'm here to help you.",
            "Welcome! Please tell me your query."
        ])

    def order_status_tool(self):
        return " Your order is confirmed and will be delivered within 3–5 working days."

    def working_hours_tool(self):
        return "our working hours are Monday to Friday, 9 AM – 6 PM."

    def refund_policy_tool(self):
        return " Refunds are processed within 7 business days after product return."

    def pricing_tool(self):
        return " Our product prices start from ₹999. Seasonal discounts are available."

    def discount_calculator_tool(self, text):
        numbers = re.findall(r'\d+', text)
        if len(numbers) >= 2:
            price = float(numbers[0])
            discount = float(numbers[1])
            final_price = price - (price * discount / 100)
            return f" Final price after {discount}% discount is ₹{final_price}"
        return "Please provide price and discount percentage."

    def complaint_tool(self):
        ticket = self.ticket_id
        self.ticket_id += 1
        return f" Your complaint has been registered successfully. Ticket ID: {ticket}"

    def feedback_tool(self):
        return " Thank you for your feedback! We appreciate your time."

    def date_time_tool(self):
        now = datetime.datetime.now()
        return f" Date: {now.strftime('%d %B %Y')} | Time: {now.strftime('%I:%M %p')}"

    def help_tool(self):
        return (
            "I can assist you with:\n"
            "- Order status\n"
            "- Pricing & discounts\n"
            "- Refund policy\n"
            "- Working hours\n"
            "- Complaints & feedback\n"
            "- Date & time\n"
            "Example: 'track my order', 'price 1000 discount 10%'"
        )

    def fallback_tool(self):
        return " I'm not sure I understood. Please try again or type 'help'."

    #  DECISION ENGINE

    def decide_action(self, user_input):
        text = user_input.lower()

        if any(word in text for word in ["hi", "hello", "hey"]):
            return self.greeting_tool()

        elif "order" in text or "track" in text:
            return self.order_status_tool()

        elif "hour" in text or "time" in text:
            return self.working_hours_tool()

        elif "refund" in text or "return" in text:
            return self.refund_policy_tool()

        elif "price" in text and "discount" in text:
            return self.discount_calculator_tool(text)

        elif "price" in text:
            return self.pricing_tool()

        elif "complaint" in text or "issue" in text:
            return self.complaint_tool()

        elif "feedback" in text:
            return self.feedback_tool()

        elif "date" in text:
            return self.date_time_tool()

        elif "help" in text:
            return self.help_tool()

        else:
            return self.fallback_tool()

    # AGENT LOOP

    def run(self):
        print(" AI Business Support Agent Activated")
        print("Type 'exit' to end the session\n")

        while True:
            user_input = input("User: ")

            if user_input.lower() == "exit":
                print("Agent: Thank you for contacting us. Have a great day!")
                break

            response = self.decide_action(user_input)
            self.memory.append({"input": user_input, "output": response})

            print("Agent:", response)

# Run the AI Agent
agent = BusinessSupportAgent()
agent.run()
```
